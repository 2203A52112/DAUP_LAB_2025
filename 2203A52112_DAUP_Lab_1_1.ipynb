```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset provided by the user
file_path = '/content/insurance.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
data_list = [data['age'], data['smoker'], data['expenses']]
```

```python
expenses_matrix = data[['age', 'expenses']].values
```

```python
# Scatter plot between age and smoking
plt.figure(figsize=(8, 5))
sns.scatterplot(x=data['age'], y=(data['smoker'] == 'yes').astype(int), marker='+', color='blue')
plt.title('Scatter Plot: Age vs Smoking')
plt.xlabel('Age')
plt.ylabel('Smoking (1 = Yes, 0 = No)')
plt.annotate('Smoker', xy=(50, 1), color='red')
plt.annotate('Non-Smoker', xy=(50, 0), color='green')
plt.grid(True)
plt.show()
```

```python
# Filter data for ages between 20 and 40
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]
```

```python
# Scatter plot between age (20 to 40) and smoking
plt.figure(figsize=(8, 5))
sns.scatterplot(x=filtered_data['age'], y=(filtered_data['smoker'] == 'yes').astype(int), marker='x', color='blue')
plt.title('Scatter Plot: Age (20-40) vs Smoking')
plt.xlabel('Age')
plt.ylabel('Smoking (1 = Yes, 0 = No)')
plt.grid(True)
plt.show()
```

```python
# Find and plot average expenses of smokers and non-smokers
expenses_by_smoking = data.groupby('smoker')['expenses'].mean().reset_index()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='smoker', y='expenses', data=expenses_by_smoking, palette='Set2')
plt.title('Average Expenses by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Average Expenses')
plt.grid(axis='y')
plt.show()
```

```python
# Number of smokers by gender
smokers_by_gender = data[data['smoker'] == 'yes'].groupby('sex').size().reset_index(name='count')
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='sex', y='count', data=smokers_by_gender, palette='muted')
plt.title('Number of Smokers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count of Smokers')
plt.grid(axis='y')
plt.show()
```

```python
# Scatter plot between age and expenses
plt.figure(figsize=(8, 5))
sns.scatterplot(x=data['age'], y=data['expenses'], color='green', alpha=0.6)
plt.title('Scatter Plot: Age vs Expenses')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.grid(True)
plt.show()
```
